# Generators and ` yield`

```python
# Sum of first n numbers:
def first_n(n):
    '''Build and return a list'''
    num, nums = 0, []
    while num < n:
        nums.append(num)
        num += 1
    return nums

print(sum(first_n(100)))
```

```python
# a generator that yields items instead of returning a list
def firstn(n):
    num = 0
    while num < n:
        yield num
        num += 1
print(sum(firstn(100)))
```

Python's `*args` and `**kwargs` constructs lets you iterate over **positional arguments** and **named arguments**: 

```python
def f(*args, **kwargs):
    print ("unnamed args: ", args)
    print ("keyword args: ", kwargs)
```

```python
f(1,2,3,a=4,b=5,c=6)
```
