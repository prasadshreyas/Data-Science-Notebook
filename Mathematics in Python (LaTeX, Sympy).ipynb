### The relationship between Golden Ratio and Fib:

What we know for Fibonacci: 

$$F_n = F_{n-1} + F_{n-2}$$

We want to study:

$$lim_{n \rightarrow \infty} \frac{F_n}{F_{n-1}}$$

So we do:

$$\frac{F_n}{F_{n-1}} = 1 + \frac{F_{n-2}}{F_{n-1}}$$

***If*** the the fraction converges to a fixed number $x$, then:

$$lim_{n \rightarrow \infty} \frac{F_n}{F_{n-1}} = lim_{n \rightarrow \infty} \frac{F_n-1}{F_{n-2}}$$

Then, to the limit:

$$\frac{F_n}{F_{n-1}} = 1 + \frac{F_{n-1}}{F_{n}}$$

$$x = 1 + \frac{1}{x}$$
